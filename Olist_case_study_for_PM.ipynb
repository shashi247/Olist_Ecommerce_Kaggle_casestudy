#Dataset Understanding

##Brazilian E-Commerce Public Dataset by Olist


1.   olist_customers_dataset.csv - Contains customer location data, i.e. zipcode, city and state. Also, it customer_id is the key to the olist_orders_dataset
2.   olist_geolocation_dataset.csv - Latitute and Longitute information against the customer zipcode (can be used for bubble map)
3. olist_order_items_dataset.csv - Contains items purchased within each order. Contains order_item_id, product_id, seller_id
4. olist_order_payments_dataset - This captures the payment method against each order, with payment_value, payment_installment and payment_type
5. olist_order_reviews_dataset - It captures the customer review against an order in the form review_score, review_comment_title, review_comment_message, creation and answer timestamp
6. olist_orders_dataset - It is the core dataset and contains customer_id, order_status and timestamps of order process like payment, approval, shipping and delivered dates
7. olist_products_dataset - It includes the data about all the product sold by olist like category and listing details like description, photos, plus the product dimensions to determine freight charges
8. olist_sellers_dataset - It includes seller information containing zipcode, city and state of the sellers
9. product_category_name_translation - Translation of product category from portugese to english

##Marketing Funnel by Olist

1. olist_closed_deals_dataset - Confirmed and approved sellers, including seller information like business_segment, lead_type, lead_behaviour_profile and gstin
2. olist_marketing_qualified_leads_dataset - MQLs, all the qualified sellers post submitting the form on the landing page

## Creating `supply_details` table

```python
# Initialize required dataframes from the dataframes dictionary
order_items_df = dataframes['olist_order_items_dataset']
products_df = dataframes['olist_products_dataset']
product_category_translation_df = dataframes['product_category_name_translation']

# Create products_with_english_category by merging products_df with translation_df
products_with_english_category = pd.merge(
    products_df[['product_id', 'product_category_name']],
    product_category_translation_df,
    on='product_category_name',
    how='left'
)

# Merge order_items_df with products_with_english_category to link sellers and prices to product categories
product_seller_price_df = pd.merge(
    order_items_df[['product_id', 'seller_id', 'price']],
    products_with_english_category[['product_id', 'product_category_name_english']],
    on='product_id',
    how='left'
)

# Rename the product_category_name_english column for clarity
product_seller_price_df = product_seller_price_df.rename(columns={'product_category_name_english': 'product_category'})

# Group by product_category to calculate the required metrics
supply_details = product_seller_price_df.groupby('product_category').agg(
    seller_count=('seller_id', 'nunique'),
    average_sales_price=('price', 'mean'),
    highest_price=('price', 'max'),
    lowest_price=('price', 'min')
).reset_index()

# Display the supply_details table
display(supply_details.head())
```

## Adding Total Revenue to `supply_details` table

```python
# Initialize orders_df as it's needed to construct order_trend
orders_df = dataframes['olist_orders_dataset']

# Construct order_trend dataframe (as it was not defined previously)
# Merge order items with orders to get purchase timestamp and order value
order_items_with_orders = pd.merge(
    order_items_df[['order_id', 'product_id', 'price']],
    orders_df[['order_id', 'order_purchase_timestamp']],
    on='order_id',
    how='left'
)

# Merge with products_with_english_category to get product category names
order_trend = pd.merge(
    order_items_with_orders,
    products_with_english_category[['product_id', 'product_category_name_english']],
    on='product_id',
    how='left'
)

# Rename columns for consistency with downstream usage, as 'order_value' and 'product_category_name' are expected.
order_trend = order_trend.rename(columns={
    'product_category_name_english': 'product_category_name',
    'price': 'order_value',
    'order_purchase_timestamp': 'order_date'
})

# Ensure 'order_date' is datetime
order_trend['order_date'] = pd.to_datetime(order_trend['order_date'])

# Calculate total revenue per product category from the order_trend dataframe
total_revenue_by_category = order_trend.groupby('product_category_name')['order_value'].sum().reset_index()
total_revenue_by_category = total_revenue_by_category.rename(columns={'product_category_name': 'product_category', 'order_value': 'total_revenue'})

# Merge total revenue into the supply_details table
supply_details = pd.merge(supply_details, total_revenue_by_category, on='product_category', how='left')

# Display the updated supply_details table
display(supply_details.head())
```

## Adding Total Revenue Per Seller to `supply_details` table

```python
# Calculate total revenue per seller
supply_details['total_revenue_per_seller'] = supply_details['total_revenue'] / supply_details['seller_count']

# Display the updated supply_details table
display(supply_details.head())
```

## Adding Total Orders to `supply_details` table

```python
# Calculate total unique orders per product category from the order_trend dataframe
total_orders_by_category = order_trend.groupby('product_category_name')['order_id'].nunique().reset_index()
total_orders_by_category = total_orders_by_category.rename(columns={'product_category_name': 'product_category', 'order_id': 'total_orders'})

# Merge total orders into the supply_details table
supply_details = pd.merge(supply_details, total_orders_by_category, on='product_category', how='left')

# Display the updated supply_details table
display(supply_details.head())
```

## Including All Product Categories in `supply_details` table

```python
# Merge order_items_df with products_with_english_category to link sellers and prices to product categories
import pandas as pd

product_seller_price_df = pd.merge(
    order_items_df[['product_id', 'seller_id', 'price']],
    products_with_english_category[['product_id', 'product_category_name_english']],
    on='product_id',
    how='left'
)

# Rename the product_category_name_english column for clarity
product_seller_price_df = product_seller_price_df.rename(columns={'product_category_name_english': 'product_category'})

# Group by product_category to calculate the required metrics
supply_details = product_seller_price_df.groupby('product_category').agg(
    seller_count=('seller_id', 'nunique'),
    average_sales_price=('price', 'mean'),
    highest_price=('price', 'max'),
    lowest_price=('price', 'min')
).reset_index()

# Calculate total revenue per product category from the order_trend dataframe
total_revenue_by_category = order_trend.groupby('product_category_name')['order_value'].sum().reset_index()
total_revenue_by_category = total_revenue_by_category.rename(columns={'product_category_name': 'product_category', 'order_value': 'total_revenue'})

# Merge total revenue into the supply_details table
supply_details = pd.merge(supply_details, total_revenue_by_category, on='product_category', how='left')

# Calculate total revenue per seller
supply_details['total_revenue_per_seller'] = supply_details['total_revenue'] / supply_details['seller_count']

# Calculate total unique orders per product category from the order_trend dataframe
total_orders_by_category = order_trend.groupby('product_category_name')['order_id'].nunique().reset_index()
total_orders_by_category = total_orders_by_category.rename(columns={'product_category_name': 'product_category', 'order_id': 'total_orders'})

# Merge total orders into the supply_details table
supply_details = pd.merge(supply_details, total_orders_by_category, on='product_category', how='left')

# Get all unique product categories from products_with_english_category
all_product_categories = products_with_english_category['product_category_name_english'].unique()
all_product_categories_df = pd.DataFrame({'product_category': all_product_categories})

# Perform a left merge to ensure all product categories are present in supply_details
# This will introduce NaN values for categories that have no sales or seller information
supply_details = pd.merge(all_product_categories_df, supply_details, on='product_category', how='left')

# --- Define growth_table and order_trend_truncated for subsequent calculations ---

# Ensure 'order_date' in order_trend is datetime and extract year, month for grouping
order_trend['order_date'] = pd.to_datetime(order_trend['order_date'])
order_trend['order_year'] = order_trend['order_date'].dt.year
order_trend['order_month'] = order_trend['order_date'].dt.month
order_trend['year_month'] = order_trend['order_date'].dt.to_period('M') # For month-on-month grouping

# For the purpose of calculating average monthly orders and revenue, we can use order_trend directly.
# Renaming for clarity as `order_trend_truncated` to match previous variable names.
order_trend_truncated = order_trend.copy()

# Aggregate monthly revenue by product category
monthly_revenue_by_category = order_trend_truncated.groupby(['product_category_name', 'year_month'])['order_value'].sum().reset_index()

# Sort for correct growth calculation
monthly_revenue_by_category = monthly_revenue_by_category.sort_values(by=['product_category_name', 'year_month'])

# Calculate month-on-month growth percentage
monthly_revenue_by_category['previous_month_revenue'] = monthly_revenue_by_category.groupby('product_category_name')['order_value'].shift(1)
monthly_revenue_by_category['order_value_growth_pct'] = (
    (monthly_revenue_by_category['order_value'] - monthly_revenue_by_category['previous_month_revenue']) /
    monthly_revenue_by_category['previous_month_revenue']
) * 100

# This is our 'growth_table'
growth_table = monthly_revenue_by_category[['product_category_name', 'year_month', 'order_value_growth_pct']].copy()
growth_table['year_month'] = growth_table['year_month'].astype(str) # Convert to string for merging/plotting

# --- New code for average month-on-month growth calculation ---

growth_table_copy = growth_table.copy()
growth_table_copy['order_year'] = pd.to_datetime(growth_table_copy['year_month']).dt.year

# Calculate average month-on-month growth for 2017
avg_growth_2017 = growth_table_copy[growth_table_copy['order_year'] == 2017].groupby('product_category_name')['order_value_growth_pct'].mean().reset_index()
avg_growth_2017 = avg_growth_2017.rename(columns={'product_category_name': 'product_category', 'order_value_growth_pct': '2017_growth%'})

# Calculate average month-on-month growth for 2018 (up to September 2018 as per dataset)
avg_growth_2018 = growth_table_copy[growth_table_copy['order_year'] == 2018].groupby('product_category_name')['order_value_growth_pct'].mean().reset_index()
avg_growth_2018 = avg_growth_2018.rename(columns={'product_category_name': 'product_category', 'order_value_growth_pct': '2018_growth%'})

# Merge 2017 growth into supply_details
supply_details = pd.merge(supply_details, avg_growth_2017, on='product_category', how='left')

# Merge 2018 growth into supply_details
supply_details = pd.merge(supply_details, avg_growth_2018, on='product_category', how='left')

# --- New code for adding average monthly orders and revenue ---

# Ensure 'order_date' is datetime (this might have already been done, but for safety in case of re-execution)
order_trend_truncated['order_date'] = pd.to_datetime(order_trend_truncated['order_date'])

# Extract year and month for grouping
order_trend_truncated['order_year'] = order_trend_truncated['order_date'].dt.year
order_trend_truncated['order_month'] = order_trend_truncated['order_date'].dt.month

# Aggregate data by product category, year, and month to get monthly totals per category
monthly_category_trend_truncated = order_trend_truncated.groupby(['product_category_name', 'order_year', 'order_month']).agg(
    monthly_orders=('order_id', 'nunique'),
    monthly_revenue=('order_value', 'sum')
).reset_index()

# Calculate the average monthly orders and revenue FOR EACH PRODUCT CATEGORY
avg_monthly_orders_per_category = monthly_category_trend_truncated.groupby('product_category_name')['monthly_orders'].mean().reset_index()
avg_monthly_revenue_per_category = monthly_category_trend_truncated.groupby('product_category_name')['monthly_revenue'].mean().reset_index()

# Merge average monthly orders into supply_details
supply_details = pd.merge(
    supply_details,
    avg_monthly_orders_per_category.rename(columns={'monthly_orders': 'month_average_orders', 'product_category_name': 'product_category'}),
    on='product_category',
    how='left'
)

# Merge average monthly revenue into supply_details
supply_details = pd.merge(
    supply_details,
    avg_monthly_revenue_per_category.rename(columns={'monthly_revenue': 'month_average_revenue', 'product_category_name': 'product_category'}),
    on='product_category',
    how='left'
)

# Display the updated supply_details table with all categories and new growth columns
display(supply_details.head())
```

**Top Performers**: Categories like watches_gifts, health_beauty, and bed_bath_table are your 'Cash Cows.' They lead in total revenue and monthly averages. watches_gifts is particularly notable for having the highest Revenue per Seller, suggesting a highly profitable segment or high-value items.

**The ASP Lever**: There is a wide variance in Average Sales Price (ASP). High-ASP categories like computers or small_appliances drive revenue with fewer units, while stationery and toys rely on high volume. As a PM, consider if your marketing spend aligns with these different unit economics.

##2. Seller Dynamics (Supply Side)

**Seller Density**: health_beauty and sports_leisure have high seller counts. This indicates a mature, competitive marketplace. Conversely, categories with high revenue but fewer sellers (like watches_gifts) might represent a 'blue ocean' or a segment where you need to protect your top-performing sellers from churn.

**Efficiency**: Monitoring Total Revenue per Seller helps identify categories where supply might be spread too thin (oversaturated) or where more sellers are needed to meet demand without diluting individual success.

##3. Growth & Seasonality Trends
**The 2017 Surge vs. 2018 Consolidation**: Many categories (e.g., art, construction_tools_safety) showed triple-digit growth in 2017. In 2018, growth rates have largely normalized.

**Identifying 'Surging' Categories**: Keep a close eye on construction_tools_safety and party_supplies. Their high MoM growth rates suggest they are emerging trends. As a PM, you should investigate if this is seasonal or a fundamental shift in user behavior.

**Dipping Categories**: security_and_services and consoles_games are seeing negative or low growth. These require a 'Fix or Exit' strategy—determine if the issue is high prices (ASP), lack of selection (Seller Count), or naturally declining interest.

##4. Seasonality Indicators
By comparing month_average_orders to total_orders, you can identify 'spiky' vs. 'stable' categories. Stable categories like perfumery provide consistent baseline revenue, whereas categories with high variance in monthly orders are likely driven by seasonal events (e.g., Black Friday, Mother’s Day).

### RFM Analysis
We will evaluate customers based on:
* **Recency (R):** How recently a customer made a purchase.
* **Frequency (F):** How often they purchase.
* **Monetary Value (M):** How much money they spend.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta

# 1. Prepare Data
# We need customer_unique_id from customers, order_purchase_timestamp from orders, and payment_value from payments
customers_df = dataframes['olist_customers_dataset']
orders_df = dataframes['olist_orders_dataset']
payments_df = dataframes['olist_order_payments_dataset']

# Join datasets
rfm_data = pd.merge(orders_df[['order_id', 'customer_id', 'order_purchase_timestamp']],
                    customers_df[['customer_id', 'customer_unique_id']], on='customer_id')
rfm_data = pd.merge(rfm_data, payments_df[['order_id', 'payment_value']], on='order_id')

# Convert to datetime
rfm_data['order_purchase_timestamp'] = pd.to_datetime(rfm_data['order_purchase_timestamp'])

# Reference date for Recency (one day after the last purchase in the dataset)
ref_date = rfm_data['order_purchase_timestamp'].max() + timedelta(days=1)

# 2. Calculate RFM Metrics
rfm = rfm_data.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (ref_date - x.max()).days,
    'order_id': 'nunique',
    'payment_value': 'sum'
}).reset_index()

rfm.columns = ['customer_id', 'recency', 'frequency', 'monetary']

# 3. Scoring (1-5, where 5 is best)
# For Recency, lower is better, so labels are reversed
rfm['R'] = pd.qcut(rfm['recency'], 5, labels=[5, 4, 3, 2, 1])
# For Frequency and Monetary, higher is better. Note: Frequency often has many 1s, so we use rank-based or handle ties
rfm['F'] = rfm['frequency'].rank(method='first').transform(lambda x: pd.qcut(x, 5, labels=[1, 2, 3, 4, 5]))
rfm['M'] = pd.qcut(rfm['monetary'], 5, labels=[1, 2, 3, 4, 5])

rfm['RFM_Score'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)

# 4. Segmenting
def segment_customer(df):
    if df['RFM_Score'] == '555':
        return 'Champions'
    elif df['R'] >= 4 and df['F'] >= 4:
        return 'Loyal Customers'
    elif df['R'] >= 3 and df['F'] >= 3:
        return 'Potential Loyalists'
    elif df['R'] >= 4 and df['F'] == 1:
        return 'New Customers'
    elif df['R'] <= 2:
        return 'At Risk / Hibernating'
    else:
        return 'Others'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)

# Display summary
segment_counts = rfm['Segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'Count']
display(segment_counts)
```

## Product Category Segmentation using Threshold-Based Approach

Given the challenges with K-Means due to the sparse and imbalanced nature of spending across many product categories, a simpler and often more interpretable approach is **threshold-based segmentation**. This involves identifying customers who are significant spenders within particular product categories.

We can define a 'Major Buyer' segment for a category if a customer's spending in that category exceeds a certain threshold (e.g., above the 75th percentile of spending for that category among all customers who bought from it).

Let's identify the top 5 product categories by overall revenue, and then for each of these top categories, define customer segments based on their spending within that category.

```python
# Identify top N product categories by overall revenue
top_n_categories = supply_details.nlargest(5, 'total_revenue')['product_category'].tolist()
print(f"Top 5 product categories by total revenue: {top_n_categories}")

# Initialize a dictionary to store segment counts for each category
category_segments = {}

for category in top_n_categories:
    # Get spending data for the current category
    category_spending = customer_category_pivot[category]

    # Filter out customers who spent 0 in this category to calculate meaningful percentiles
    active_category_spenders = category_spending[category_spending > 0]

    if not active_category_spenders.empty:
        # Define a threshold (e.g., 75th percentile of spending for this category)
        threshold = active_category_spenders.quantile(0.75)

        # Segment customers for this category
        major_buyers = active_category_spenders[active_category_spenders >= threshold].index
        other_buyers = active_category_spenders[active_category_spenders < threshold].index

        category_segments[category] = {
            'Major Buyers': len(major_buyers),
            'Other Buyers': len(other_buyers),
            'Threshold_75th_Percentile': threshold
        }
    else:
        category_segments[category] = {
            'Major Buyers': 0,
            'Other Buyers': 0,
            'Threshold_75th_Percentile': 0
        }

# Display the segmentation results for top categories
print("\nCustomer Segmentation for Top Product Categories (Threshold-Based):")
for category, data in category_segments.items():
    print(f"\nCategory: {category}")
    print(f"  - Major Buyers (spending >= {data['Threshold_75th_Percentile']:.2f}): {data['Major Buyers']} customers")
    print(f"  - Other Buyers (spending < {data['Threshold_75th_Percentile']:.2f} & > 0): {data['Other Buyers']} customers")
    print(f"  - Non-Buyers: {len(customer_category_pivot) - (data['Major Buyers'] + data['Other Buyers'])} customers")

# Example: Assigning a 'Primary Category' based on highest spend
# This creates a segment where each customer is assigned to the category they spent most in
customer_category_pivot['Primary_Category_Segment'] = customer_category_pivot.idxmax(axis=1)

print("\nDistribution of customers by their Primary Category Segment (category of highest spend):")
display(customer_category_pivot['Primary_Category_Segment'].value_counts().head(10))
```

## Interpreting Threshold-Based Product Category Segments

This threshold-based approach provides more direct and actionable insights into customer preferences for specific product categories:

*   **Major Buyers**: For each of the top product categories, we've identified customers who are significant spenders within that category (e.g., those above the 75th percentile of spending). These customers are highly engaged and valuable for category-specific marketing.

*   **Other Buyers**: These are customers who have purchased from the category but are not in the top spending tier. They could be targeted with upselling or cross-selling strategies.

*   **Non-Buyers**: A large segment of customers who have not purchased from a given category. These represent potential expansion opportunities.

*   **Primary Category Segment**: By assigning each customer to the product category in which they spent the most, we get a clear picture of their primary interest. This allows for highly personalized recommendations and marketing efforts.

This method avoids the issues of K-Means with sparse data and provides clear, interpretable segments. You can now tailor marketing campaigns, product development, and inventory management based on these defined customer groups and their specific category preferences.
