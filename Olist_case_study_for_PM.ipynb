#Dataset Understanding

##Brazilian E-Commerce Public Dataset by Olist


1.   olist_customers_dataset.csv - Contains customer location data, i.e. zipcode, city and state. Also, it customer_id is the key to the olist_orders_dataset
2.   olist_geolocation_dataset.csv - Latitute and Longitute information against the customer zipcode (can be used for bubble map)
3. olist_order_items_dataset.csv - Contains items purchased within each order. Contains order_item_id, product_id, seller_id
4. olist_order_payments_dataset - This captures the payment method against each order, with payment_value, payment_installment and payment_type
5. olist_order_reviews_dataset - It captures the customer review against an order in the form review_score, review_comment_title, review_comment_message, creation and answer timestamp
6. olist_orders_dataset - It is the core dataset and contains customer_id, order_status and timestamps of order process like payment, approval, shipping and delivered dates
7. olist_products_dataset - It includes the data about all the product sold by olist like category and listing details like description, photos, plus the product dimensions to determine freight charges
8. olist_sellers_dataset - It includes seller information containing zipcode, city and state of the sellers
9. product_category_name_translation - Translation of product category from portugese to english

##Marketing Funnel by Olist

1. olist_closed_deals_dataset - Confirmed and approved sellers, including seller information like business_segment, lead_type, lead_behaviour_profile and gstin
2. olist_marketing_qualified_leads_dataset - MQLs, all the qualified sellers post submitting the form on the landing page

**The ASP Lever**: There is a wide variance in Average Sales Price (ASP). High-ASP categories like computers or small_appliances drive revenue with fewer units, while stationery and toys rely on high volume. As a PM, consider if your marketing spend aligns with these different unit economics.

##2. Seller Dynamics (Supply Side)

**Seller Density**: health_beauty and sports_leisure have high seller counts. This indicates a mature, competitive marketplace. Conversely, categories with high revenue but fewer sellers (like watches_gifts) might represent a 'blue ocean' or a segment where you need to protect your top-performing sellers from churn.

**Efficiency**: Monitoring Total Revenue per Seller helps identify categories where supply might be spread too thin (oversaturated) or where more sellers are needed to meet demand without diluting individual success.

##3. Growth & Seasonality Trends
**The 2017 Surge vs. 2018 Consolidation**: Many categories (e.g., art, construction_tools_safety) showed triple-digit growth in 2017. In 2018, growth rates have largely normalized.

**Identifying 'Surging' Categories**: Keep a close eye on construction_tools_safety and party_supplies. Their high MoM growth rates suggest they are emerging trends. As a PM, you should investigate if this is seasonal or a fundamental shift in user behavior.

**Dipping Categories**: security_and_services and consoles_games are seeing negative or low growth. These require a 'Fix or Exit' strategy—determine if the issue is high prices (ASP), lack of selection (Seller Count), or naturally declining interest.

##4. Seasonality Indicators
By comparing month_average_orders to total_orders, you can identify 'spiky' vs. 'stable' categories. Stable categories like perfumery provide consistent baseline revenue, whereas categories with high variance in monthly orders are likely driven by seasonal events (e.g., Black Friday, Mother’s Day).
